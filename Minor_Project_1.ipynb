Project Option 1: Console- based Smart Calculator


```python
import numpy as np
print("Enter 1 for addition.")#Guideline for the users to enter the no.s according to which operation they want to perform
print("Enter 2 for subtraction.")
print("Enter 3 for product.")
print("Enter 4 for division.")
print("Enter 5 for modulus.")
print("Enter 6 for power calculation.")
ch=int(input("Enter no. according to which operation you want to do"))#taking no as which operation they want to perform
if ch==1:
  lst=[]
  n=int(input("Enter How many no.s to calculate"))
  for i in range(n):
     value=int(input("Enter value:"))
     lst.append(value)
  print(lst)
  add=sum(lst)
  print(add)
elif ch==2:
  a=int(input("Enter first no."))
  b=int(input("Enter second no."))
  diff=np.subtract(a,b)
  print(diff)
elif ch==3:
  lst=[]
  n=int(input("Enter How many no.s to calculate"))
  for i in range(n):
     value=int(input("Enter value:"))
     lst.append(value)
  print(lst)
  mul=np.prod(lst)
  print(mul)
elif ch==4:
    a=int(input("Enter first no."))
    b=int(input("Enter second no."))
    if b!=0:
      div=np.divide(a,b)
      print(div)
    else:
      print("Division by zero is undefined in maths")
elif ch==5:
  a=int(input("Enter first no."))
  b=int(input("Enter second no."))
  modulus=np.mod(a,b)
  print(modulus)
else:
  a=int(input("Enter the base"))
  b=int(input("Enter the power"))
  po=np.power(a,b)
  print(po)
```

```python
import numpy as np
ns=int(input("Enter the no. of students"))
lm=[]
ls=[]
le=[]
for i in range(ns):
  mm=int(input(f"Enter the Maths Marks of student{i+1}"))
  lm.append(mm)
  sm=int(input(f"Enter the Science Marks of student{i+1}"))
  ls.append(sm)
  em=int(input(f"Enter the English Marks of student{i+1}"))
  le.append(em)
  TM=mm+sm+em
  AM=(mm+sm+em)/3
  if AM>85:
    print(f"Student {i+1} got A grade")
  elif 69<AM<85:
    print(f"Student{i+1} B grade")
  elif 49<AM<69:
    print(f"Student {i+1} C grade")
  else:
    print(f"Student {i+1} failed")


  print(f"The total marks is{TM}")
  print(f"The average marks is{AM}")
print(f"List of maths marks of every student{np.array(lm)}")
print(f"List of science marks of every student{np.array(ls)}")
print(f"List of english marks of every student{np.array(le)}")

if mm>em and mm>sm:
  print("Maximum Students got highest marks in Maths.")
elif sm>mm and sm>em:
  print("Maximum Students got highest marks in Science.")
elif em>mm and em>sm:
  print("Maximum Students got highest marks in English.")
elif em==sm:
  print("Maximum Students got highest marks in English and Science.")
elif em==mm:
  print("Maximum Students got highest marks in English and Maths.")
elif mm==sm:
  print("Maximum Students got highest marks in Maths and science")
else:
  print("Maximum Students got equal marks in every subject.")


```
